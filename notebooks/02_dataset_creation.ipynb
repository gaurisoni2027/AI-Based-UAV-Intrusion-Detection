# Dataset Creation

## Objective
Create a clean network traffic dataset from the original IEEE T-ITS dataset for building an intrusion detection system (IDS) for UAV communication networks.

## Why a New Dataset?
- The original dataset contains both network traffic and UAV telemetry data.
- It has schema inconsistencies and missing labels.
- A separate dataset will make preprocessing, experimentation, and model development more reliable and reproducible.

## Output
This notebook will generate:

`data/interim/network_dataset.csv`

which will contain only the records relevant to network-based intrusion detection.

## Approach

1. Load the original dataset.
2. Identify rows corresponding to network traffic.
3. Remove corrupted or irrelevant records.
4. Preserve all network-related features and the target label.
5. Save the cleaned network dataset for further analysis.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/dataset.csv", low_memory=False)

print("Original Shape:", df.shape)
df.head()
```

```python
header_rows = df["class"] == "class"

print("Repeated header rows:", header_rows.sum())

df = df[~header_rows].copy()

print("Shape after removing header rows:", df.shape)
```

# Dataset Creation Decision

The original dataset contains heterogeneous data modalities, including network traffic records and UAV telemetry measurements.

Since the objective of this work is intrusion detection in UAV communication networks, only network traffic records will be retained.

Telemetry and other schema-inconsistent records will be excluded to ensure:

- consistent feature semantics,
- reliable preprocessing,
- meaningful feature analysis,
- reproducible experiments.

The excluded data may be investigated in future work for developing a multimodal cyber-physical intrusion detection system.

```python
#keeping only the valid labels 
network_df = df[df["class"].notna()].copy()
```

## Creating the Intermediate Dataset

A new intermediate dataset (`network_dataset.csv`) is created by retaining only samples with valid attack labels.

Reasons for creating an intermediate dataset:

- The original dataset contains heterogeneous data modalities.
- Unlabeled telemetry and FDI records are excluded from the initial study.
- The resulting dataset contains only labeled network traffic samples relevant to supervised intrusion detection.

This intermediate dataset will be used for all subsequent preprocessing, exploratory data analysis, and model development.

```python
network_df.to_csv(
    "../data/interim/network_dataset.csv",
    index=False
)
```
